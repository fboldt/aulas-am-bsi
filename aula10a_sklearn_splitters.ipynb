```python
from sklearn.datasets import load_wine
X, y = load_wine(return_X_y=True)
print(X.shape)
```

```python
import numpy as np
unique_classes, counts = np.unique(y, return_counts=True)
for cls, count in zip(unique_classes, counts):
    print(f"Class {cls}: {count} examples")
```

```python
import numpy as np

def print_class_percentages(y):
    unique_classes, counts = np.unique(y, return_counts=True)
    total_examples = len(y)
    print("Class Percentages:")
    for cls, count in zip(unique_classes, counts):
        percentage = (count / total_examples) * 100
        print(f"Class {cls}: {percentage:.2f}%")

# Example usage with the 'y' variable from the previous steps:
print_class_percentages(y)
```

#holdout

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print("Train")
print_class_percentages(y_train)
print("\nTest")
print_class_percentages(y_test)
```

#stratified holdout

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    stratify=y)
print("Train")
print_class_percentages(y_train)
print("\nTest")
print_class_percentages(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
model = KNeighborsClassifier()
```

#KFold

```python
from sklearn.model_selection import KFold
splitter = KFold(n_splits=5)
def perform_evaluation(splitter, X, y):
  scores = []
  for train_index, test_index in splitter.split(X):
    model.fit(X[train_index], y[train_index])
    y_pred = model.predict(X[test_index])
    score = accuracy_score(y[test_index], y_pred)
    scores.append(score)
    print_class_percentages(y[test_index])
    print()
  print(scores)

perform_evaluation(splitter, X, y)
```

#Shuffled KFold

```python
from sklearn.model_selection import KFold
splitter = KFold(n_splits=5, shuffle=True)
perform_evaluation(splitter, X, y)
```

#Stratified KFold

```python
from sklearn.model_selection import StratifiedKFold
splitter = StratifiedKFold(n_splits=5)
def perform_stratified_evaluation(splitter, X, y):
  scores = []
  for train_index, test_index in splitter.split(X,y):
    model.fit(X[train_index], y[train_index])
    y_pred = model.predict(X[test_index])
    score = accuracy_score(y[test_index], y_pred)
    scores.append(score)
    print_class_percentages(y[test_index])
    print()
  print(scores)
perform_stratified_evaluation(splitter, X, y)
```

# Stratified Shuffled KFold

```python
from sklearn.model_selection import StratifiedKFold
splitter = StratifiedKFold(n_splits=5, shuffle=True)
perform_stratified_evaluation(splitter, X, y)
```

#Leave one out

```python
from sklearn.model_selection import LeaveOneOut
splitter = LeaveOneOut()
scores = []
for train_index, test_index in splitter.split(X):
  model.fit(X[train_index], y[train_index])
  y_pred = model.predict(X[test_index])
  score = accuracy_score(y[test_index], y_pred)
  scores.append(score)
print(len(scores))
print(f"Accuracy: {np.sum(scores)/len(scores)}")
```

#Repeated KFold

```python
from sklearn.model_selection import RepeatedKFold
splitter = RepeatedKFold(n_splits=3, n_repeats=2)
for i, (train_index, test_index) in enumerate(splitter.split(X)):
  print(f"Fold {i+1}")
  print_class_percentages(y[test_index])
```

#Repeated Stratified KFold

```python
from sklearn.model_selection import RepeatedStratifiedKFold
splitter = RepeatedStratifiedKFold(n_splits=3, n_repeats=2)
for i, (train_index, test_index) in enumerate(splitter.split(X, y)):
  print(f"Fold {i+1}")
  print_class_percentages(y[test_index])
```

# Grouped KFold

```python
import numpy as np
from sklearn.model_selection import GroupShuffleSplit
X_tmp = np.ones(shape=(8, 2))
y_tmp = np.ones(shape=(8, 1))
groups = np.array([1, 1, 2, 2, 2, 3, 3, 3])
print(groups.shape)
gss = GroupShuffleSplit(n_splits=len(set(groups)), train_size=.7, random_state=42)
gss.get_n_splits()
print(gss)
for i, (train_index, test_index) in enumerate(gss.split(X_tmp, y_tmp, groups)):
    print(f"Fold {i}:")
    print(f"  Train: index={train_index}, group={groups[train_index]}")
    print(f"  Test:  index={test_index}, group={groups[test_index]}")
```
