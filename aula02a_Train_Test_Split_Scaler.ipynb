```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
print(housing.data.shape, housing.target.shape)
print(housing.feature_names[0:6])
```

```python
housing = fetch_california_housing(as_frame=True)
df = housing.frame
df
```

```python
import seaborn as sns
corr = df.corr()
sns.heatmap(corr, annot=True)
```

```python
X = df['MedInc'].values.reshape(-1, 1)
y = df['MedHouseVal'].values
print(X.shape, y.shape)
```

```python
import numpy as np

def train_test_slit(X, y, test_size=0.2, random_state=42):
    np.random.seed(random_state)
    indices = np.random.permutation(len(X))
    test_size = int(len(X) * test_size)
    test_indices = indices[:test_size]
    train_indices = indices[test_size:]
    return X[train_indices], X[test_indices], y[train_indices], y[test_indices]

X_train, X_test, y_train, y_test = train_test_slit(X, y, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
print(X_train[:5].T)
print(X_test[:5].T)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_scatter(X, y):
  plt.figure(figsize=(10, 6))
  sns.scatterplot(x=X.flatten(), y=y, alpha=0.6)
  plt.title('Scatter Plot of Median Income vs. Median House Value')
  plt.xlabel('Median Income')
  plt.ylabel('Median House Value')
  plt.grid(True)
  plt.show()

plot_scatter(X_train, y_train)
```

```python
def max_min_scaler(X):
    min_val = X.min()
    max_val = X.max()
    return (X - min_val) / (max_val - min_val)

X_train_scaled = max_min_scaler(X_train)

plot_scatter(X_train_scaled, y_train)
```

```python
def max_min_scaler(X, min=-1, max=1):
    min_val = X.min()
    max_val = X.max()
    return min + (X - min_val) * (max - min) / (max_val - min_val)

X_train_scaled = max_min_scaler(X_train)

plot_scatter(X_train_scaled, y_train)
```

```python
def standard_scaler(X):
    mean = X.mean()
    std = X.std()
    return (X - mean) / std

X_train_scaled = standard_scaler(X_train)

plot_scatter(X_train_scaled, y_train)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class StandardScaler(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
      self.mean = X.mean()
      self.std = X.std()
      return self
    def transform(self, X, y=None):
      return (X - self.mean) / self.std

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

plot_scatter(X_train_scaled, y_train)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_train)
lin_reg.score(X_test_scaled, y_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_train_pred = lin_reg.predict(X_train_scaled)

plt.figure(figsize=(10, 7))
plt.scatter(X_train, y_train, alpha=0.6, label='Actual MedHouseVal')
plt.scatter(X_train, y_train_pred, alpha=0.6, label='Predicted MedHouseVal', color='red')

plt.title('Actual vs. Predicted Median House Value by Median Income (train Set)')
plt.xlabel('Median Income (MedInc)')
plt.ylabel('Median House Value (MedHouseVal)')
plt.legend()
plt.grid(True)
plt.show()
```
