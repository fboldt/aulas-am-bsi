```python
from sklearn.datasets import load_wine
import numpy as np

X, y = load_wine(return_X_y=True)
print(X.shape)
print()
```

```python
unique_labels, counts = np.unique(y, return_counts=True)
print("Label counts:")
for label, count in zip(unique_labels, counts):
    print(f"Label {label}: {count}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

# Conseguindo recall alto

```python
positivos = 10
negativos = 990
positivos_porcentagem = positivos / (positivos + negativos) * 100
print(f"Porcentagem de positivos: {positivos_porcentagem:.2f}%")
negativos_porcentagem = negativos / (positivos + negativos) * 100
print(f"Porcentagem de negativos: {negativos_porcentagem:.2f}%")

############
# 10 # 0   #
# 90 # 900 #
############

verdadeiro_positivo = 10
vedadeiro_negativo = 900
falso_positivo = 90
falso_negativo = 0

print()

accuracia = (verdadeiro_positivo + vedadeiro_negativo) / (positivos + negativos)
print(f"Acurácia: {accuracia}")

precisao = verdadeiro_positivo / (verdadeiro_positivo + falso_positivo)
print(f"Precisão: {precisao}")

recall = verdadeiro_positivo / positivos
print(f"Recall: {recall}")
```

# Conseguindo precisão alta às custas de recall baixo

```python
positivos = 100
negativos = 900
positivos_porcentagem = positivos / (positivos + negativos) * 100
print(f"Porcentagem de positivos: {positivos_porcentagem:.2f}%")
negativos_porcentagem = negativos / (positivos + negativos) * 100
print(f"Porcentagem de negativos: {negativos_porcentagem:.2f}%")

############
# 1  # 99  #
# 0  # 900 #
############

verdadeiro_positivo = 1
vedadeiro_negativo = 900
falso_positivo = 0
falso_negativo = 99

print()

accuracia = (verdadeiro_positivo + vedadeiro_negativo) / (positivos + negativos)
print(f"Acurácia: {accuracia}")

precisao = verdadeiro_positivo / (verdadeiro_positivo + falso_positivo)
print(f"Precisão: {precisao}")

recall = verdadeiro_positivo / positivos
print(f"Recall: {recall}")
```

# F1-score

```python
# Média harmônica da precisão e da revocação

f1_score = (2 * precisao * recall)/(precisao + recall)
print(f"F1-score: {f1_score}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score

print(classification_report(y_test, y_pred))
print(f"Acurácia: {accuracy_score(y_test, y_pred):.2f}")
print(f"Precisão: {precision_score(y_test, y_pred, average='macro'):.2f}")
print(f"Recall: {recall_score(y_test, y_pred, average='macro'):.2f}")
print(f"F1-score: {f1_score(y_test, y_pred, average='macro'):.2f}")
print(f"F-beta = 0.1: {fbeta_score(y_test, y_pred, beta=0.1, average='macro'):.5f}")
print(f"F-beta = 1.0: {fbeta_score(y_test, y_pred, beta=1.0, average='macro'):.5f}")
print(f"F-beta = 100: {fbeta_score(y_test, y_pred, beta=100, average='macro'):.5f}")
```
